### Imports

```python
from pathlib import Path
import pandas as pd
```

### Load Data

```python
base_dir = Path.cwd()
out_dir = base_dir / "results"
df = pd.read_csv(base_dir/"final.csv")
df.head()
```

### Main

```python
df = df.iloc[:,1:].copy()
df.drop_duplicates(inplace=True)
df.dropna(inplace=True)
df = df[df['Ratio(Tumor Count/Blood Count)']!=0]
df = df.sort_values('Ratio(Tumor Count/Blood Count)',ascending=False).copy()
df.reset_index(drop=True,inplace=True)
```

```python
df.value_counts('Project ID')
```

```python
df.to_csv(out_dir/"EGFR.csv",index=False)
```
