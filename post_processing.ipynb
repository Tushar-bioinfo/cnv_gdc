### Imports

```python
from pathlib import Path
import pandas as pd
```

### Load Data

```python
out_dir = Path("/Users/tusharsingh/Library/CloudStorage/OneDrive-UniversityofSouthFlorida/JOB/cnv/cnv_tcga/results/run2")
df = pd.read_csv("/Users/tusharsingh/Library/CloudStorage/OneDrive-UniversityofSouthFlorida/JOB/cnv/cnv_tcga/results/run2/final.csv")
df.head()
```

### Main

```python
df = df.iloc[:,1:].copy()
df.drop_duplicates(inplace=True)
df.dropna(inplace=True)
df = df[df['Ratio(Tumor Count/Blood Count)']!=0]
df = df.sort_values('Ratio(Tumor Count/Blood Count)',ascending=False).copy()
df.reset_index(drop=True,inplace=True)
```

```python
df.value_counts('Project ID')
```

```python
types = df['Project ID'].value_counts().index.to_list()
```

```python
writer = pd.ExcelWriter(out_dir / f'Mien1.xlsx', engine='xlsxwriter')


for i in types:
    temp = df[df['Project ID']==i].copy()
    temp.reset_index(drop=True,inplace=True)
    temp.to_excel(writer, sheet_name=i, index=False)

writer.close()
```
